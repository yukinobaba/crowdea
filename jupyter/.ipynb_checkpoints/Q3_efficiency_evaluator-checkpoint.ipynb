```python
%matplotlib inline
import matplotlib.pyplot as plt

from pathlib import Path
import pandas as pd
import numpy as np
import collections
import scipy.stats

import warnings
warnings.filterwarnings('ignore')
```

```python
from ranking import RankingMeasures
```

```python
data_cats = ['idea', 'design']
#data_keys = {'idea': ['bike', 'cheat', 'festival', 'meeting', 'name', 'night', 'trash', 'visitor'], 
#             'design': ['ai_character', 'olympic']}
data_keys = {'idea': ['bike', 'cheat', 'meeting', 'night', 'visitor'], 
             'design': ['ai_character', 'olympic']}
data_keys_arr = data_keys['idea'] + data_keys['design']
```

```python
num_evaluators_arr = [20, 50, 100]
d_arr = [2, 3]
methods = ['crowdea_%d' % d for d in d_arr] + ['crowdbt', 'bt']
```

```python
data_dir = Path('../data/')

ndcg5_data = []
ndcg10_data = []

for data_cat in data_cats:
    for data_key in data_keys[data_cat]:
        truth_df = pd.read_csv(data_dir / data_cat / data_key / 'truth.tsv', index_col=0, sep='\t')        
        
        for num_evaluators in num_evaluators_arr:
            if data_key == 'olympic' and num_evaluators == 100:
                continue
                
            result_dir = Path('../result_e%d/' % num_evaluators)                        
            
            for method in methods:                
                for trial in range(10):   
                    result_dir_sub = result_dir / data_cat / data_key / str(trial)
                    
                    if method == 'bt' or method == 'crowdbt':
                        for param in [0.001, 0.01, 0.1]:
                            if method == 'bt':
                                p = np.loadtxt(result_dir_sub / 'bt_x_beta{}.dat'.format(param))
                            elif method == 'crowdbt':
                                p = np.loadtxt(result_dir_sub / 'crowdbt_x_lambda{}.dat'.format(param))

                            for viewpoint in truth_df.columns:
                                if viewpoint[0] != '*': 
                                    continue

                                p_truth = truth_df.loc[:, viewpoint].values
                                rm = RankingMeasures(p, p_truth)
                                ndcg5 = rm.nDCG(5)
                                ndcg10 = rm.nDCG(10)

                                ndcg5_data.append({'data_key': data_key, 'ndcg5': ndcg5, 
                                                   'method': method, 'viewpoint': viewpoint, 
                                                   'trial': trial, 'num_evaluators': num_evaluators, 
                                                   'param': param})
                                ndcg10_data.append({'data_key': data_key, 'ndcg10': ndcg10, 
                                                   'method': method, 'viewpoint': viewpoint, 
                                                   'trial': trial, 'num_evaluators': num_evaluators, 
                                                   'param': param})
                            
                    else:
                        d = int(method.split('_')[-1])                        
                        x = np.loadtxt(result_dir_sub / 'crowdea_x_alpha0.1_d{}.dat'.format(d))

                        for viewpoint in truth_df.columns:
                            if viewpoint[0] != '*': 
                                continue

                            p_truth = truth_df.loc[:, viewpoint].values
                            v_opt = np.linalg.lstsq(x, p_truth)[0]
                            p = x.dot(v_opt)

                            rm = RankingMeasures(p, p_truth)
                            ndcg5 = rm.nDCG(5)
                            ndcg10 = rm.nDCG(10)

                            ndcg5_data.append({'data_key': data_key, 'ndcg5': ndcg5, 
                                               'method': method, 'viewpoint': viewpoint, 
                                               'trial': trial, 'num_evaluators': num_evaluators})
                            ndcg10_data.append({'data_key': data_key, 'ndcg10': ndcg10, 
                                               'method': method, 'viewpoint': viewpoint, 
                                               'trial': trial, 'num_evaluators': num_evaluators}) 
                    
score_dfs = {'ndcg5': pd.DataFrame(ndcg5_data),
            'ndcg10': pd.DataFrame(ndcg10_data)}
```

```python
# select best parameters of BT and CrowdBT for each (measure, data_key)
best_params = collections.defaultdict(dict)

for measure in score_dfs.keys():
    score_df = score_dfs[measure]
    for method in ['bt', 'crowdbt']:
        avg_df = score_df.loc[score_df.method == method].groupby(['data_key', 'method', 'num_evaluators', 'trial', 'param']).mean()
        for data_key in data_keys_arr:
            for num_evaluators in num_evaluators_arr:
                if data_key == 'olympic' and num_evaluators == 100:
                    continue
                for trial in range(10):
                    tmp_df = avg_df.loc[(data_key, method, num_evaluators, trial)]
                    tmp_df = tmp_df.sample(tmp_df.shape[0]) # for tie-breaking
                    best_params[measure][(method, data_key, num_evaluators, trial)] = tmp_df.sort_values(measure, ascending=False).index[0]
            
for measure in score_dfs.keys():
    score_df = score_dfs[measure]
    for index, row in score_df.iterrows():
        if row.method in ['bt', 'crowdbt']:
            if best_params[measure][(row.method, row.data_key, row.num_evaluators, row.trial)] != row.param:
                score_df.loc[index, measure] = np.nan
    score_df.drop('param', axis=1, inplace=True)
    score_df = score_df.dropna()
    score_dfs[measure] = score_df
```

```python
avg_dfs = {}
for measure in score_dfs.keys():
    avg_dfs[measure] = score_dfs[measure].groupby(['data_key', 'method', 'num_evaluators'])[measure].mean()
```

```python
std_dfs = {}
for measure in score_dfs.keys():
    std_dfs[measure] = score_dfs[measure].groupby(['data_key', 'method', 'num_evaluators', 'trial'])[measure].mean().groupby(['data_key', 'method', 'num_evaluators']).std()
```

```python
result_dfs = {}
std_result_dfs = {}

for measure in score_dfs.keys():
    for method in methods:    
        result_df = pd.DataFrame(index=data_keys_arr, 
                                 columns=['#evaluators=%d' % n for n in num_evaluators_arr])
        std_result_df = pd.DataFrame(index=data_keys_arr, 
                                 columns=['#evaluators=%d' % n for n in num_evaluators_arr])
        for data_key in data_keys_arr:
            for num_evaluators in num_evaluators_arr:
                if data_key == 'olympic' and num_evaluators == 100:
                    v = np.nan
                    e = np.nan
                else:
                    v = avg_dfs[measure].loc[(data_key, method, num_evaluators)]
                    e = std_dfs[measure].loc[(data_key, method, num_evaluators)]
                    
                result_df.loc[data_key, '#evaluators=%d' % num_evaluators] = v
                std_result_df.loc[data_key, '#evaluators=%d' % num_evaluators] = e
        result_dfs[(measure, method)] = result_df
        std_result_dfs[(measure, method)] = std_result_df
```

```python
data_key_tex = {}
for data_key in data_keys_arr:
    data_key_tex[data_key] = data_key.capitalize()
    if data_key == 'ai_character':
        data_key_tex[data_key] = 'Character'
```

```python
measure = 'ndcg10'
lines = []

s = []
for d in d_arr:
    s.append('c|' * (len(num_evaluators_arr)))
    
lines.append('\\begin{tabular}{|l||%s}' % ('|'.join(s)))
lines.append('\\hline')

#header1
elems = ['']
for d in d_arr:
    if d == d_arr[-1]:
        elems.append('\\multicolumn{%d}{c|}{\\textsc{CrowDEA} (d=%d)}' % (len(num_evaluators_arr), d))
    else:
        elems.append('\\multicolumn{%d}{c||}{\\textsc{CrowDEA} (d=%d)}' % (len(num_evaluators_arr), d))
        
lines.append('&'.join(elems) + '\\\\')

lines.append('\\cline{1-2}')
lines.append('\\cline{2-4}')
lines.append('\\cline{5-7}')
lines.append('\\cline{8-10}')
lines.append('\\cline{11-13}')

#header2
elems = ['\\#evaluators']
for d in d_arr:
    for num_evaluators in num_evaluators_arr:
        elems.append('$%d$' % num_evaluators)
lines.append('&'.join(elems) + '\\\\')

lines.append('\\hline')
lines.append('\\hline')

#contents
for data_key in data_keys_arr:
    elems = [data_key_tex[data_key]]
    for d in d_arr:
        for num_evaluators in num_evaluators_arr:        
            (v, e) = result_dfs[(measure, 'crowdea_%d' % d)].loc[data_key, '#evaluators=%d' % num_evaluators]
            if np.isnan(v):
                elems.append('-')
            else:
                elems.append('$%.3f \pm %.3f$' % (v, e))
                
    lines.append('&'.join(elems) + '\\\\')
    if data_key == 'visitor':
        lines.append('\\hline')
        
lines.append('\\hline')
lines.append('\\end{tabular}')

print('\n'.join(lines))
```

```python
measure = 'ndcg10'
tmp_methods = ['bt', 'crowdbt', 'crowdea_2']
lines = []

s = []
for _ in tmp_methods:
    s.append('c|' * (len(num_evaluators_arr)))
    
lines.append('\\begin{tabular}{|l||%s}' % ('|'.join(s)))
lines.append('\\hline')

#header1
elems = ['']
elems.append('\\multicolumn{%d}{c||}{\\textsc{BT}}' % (len(num_evaluators_arr)))
elems.append('\\multicolumn{%d}{c||}{\\textsc{CrowdBT}}' % (len(num_evaluators_arr)))
elems.append('\\multicolumn{%d}{c|}{\\textsc{CrowDEA} (d=%d)}' % (len(num_evaluators_arr), 2))
lines.append('&'.join(elems) + '\\\\')

lines.append('\\cline{1-2}')
lines.append('\\cline{2-4}')
lines.append('\\cline{5-7}')
lines.append('\\cline{8-10}')
lines.append('\\cline{11-13}')

#header2
elems = ['\\#evaluators']
for _ in tmp_methods:
    for num_evaluators in num_evaluators_arr:
        elems.append('$%d$' % num_evaluators)
lines.append('&'.join(elems) + '\\\\')

lines.append('\\hline')
lines.append('\\hline')

#contents
for data_key in data_keys_arr:
    elems = [data_key_tex[data_key]]
    for method in tmp_methods:
        for num_evaluators in num_evaluators_arr:        
            v = result_dfs[(measure, method)].loc[data_key, '#evaluators=%d' % num_evaluators]
            if np.isnan(v):
                elems.append('-')
            else:
                elems.append('$%.3f$' % (v))
                
    lines.append('&'.join(elems) + '\\\\')
    if data_key == 'visitor':
        lines.append('\\hline')
        
lines.append('\\hline')
lines.append('\\end{tabular}')

print('\n'.join(lines))
```

```python
measure = 'ndcg10'

plt.figure(figsize=(16, 5))
plt.style.use('seaborn-ticks')
plt.rcParams["font.size"] = 16

x = num_evaluators_arr

for i, data_key in enumerate(data_keys_arr):
    plt.subplot(2, 4, i+1)
    
    for method in methods:
        y = [result_dfs[(measure, method)].loc[data_key, '#evaluators=%d' % num_evaluators] for num_evaluators in num_evaluators_arr]
        e = [std_result_dfs[(measure, method)].loc[data_key, '#evaluators=%d' % num_evaluators] for num_evaluators in num_evaluators_arr]
        label = label_dict[method]
        if marker_dict[method] == '.':
            plt.errorbar(x, y, yerr=e, capsize=2, elinewidth=1, fmt='%s-' % marker_dict[method], color=color_dict[method], label=label, lw=3, ms=8, mew=1.5, mec=color_dict[method])
        else:
            plt.errorbar(x, y, yerr=e, capsize=2, elinewidth=1, fmt='%s-' % marker_dict[method], color=color_dict[method], label=label, lw=3, ms=8, mew=1.5, mec=color_dict[method], mfc='white')

    plt.title(data_key_tex[data_key])    
    plt.ylim(0.7, 0.95)
    plt.xlabel('Num. of evaluators', labelpad=2)
    plt.ylabel('nDCG@10', labelpad=2)
    if data_key == 'olympic':
        plt.xticks([20, 50, 100]) 
    else:
        plt.xticks([20, 50, 100]) 
    
anchor_y = -0.7
#plt.legend(loc="lower center", bbox_to_anchor=(-0.8, anchor_y), ncol=4)
plt.tight_layout(pad=0., w_pad=0.5, h_pad=0.5)
plt.savefig('fig_efficiency.pdf', bbox_inches="tight", pad_inches=0.)
```
